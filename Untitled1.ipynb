# Data cleaning


```python
import pandas as pd
```

```python
df_hist = pd.read_csv('fifa_worldcup_matches.csv')
df_fix = pd.read_csv('fifa_worldcup_fixture.csv')
df_missing = pd.read_csv('fifa_worldcup_missing_data.csv')
```

```python
df_fix['home'] = df_fix['home'].str.strip()
```

```python
df_fix['away'] =df_fix['away'].str.strip()
```

```python
df_fix
```

```python
df_missing
```

```python
df_missing[df_missing['home'].isnull()]
```

```python
df_missing.dropna(inplace=True)
```

```python
df_hist= pd.concat([df_hist, df_missing], ignore_index=True)
df_hist.drop_duplicates(inplace=True)
df_hist.sort_values('year', inplace=True)
```

```python
df_hist
```

```python
delete= df_hist[df_hist['home'].str.contains('Sweden') & df_hist['away'].str.contains('Austria')].index
```

```python
df_hist.drop(index=delete, inplace= True)
```

```python
 df_hist[df_hist['home'].str.contains('Sweden') & df_hist['away'].str.contains('Austria')]
```

```python
df_hist[df_hist['score'].str.contains('[^\d–]')]
```

```python
df_hist['score'].str.replace('[^\d–]', '', regex= True)
```

```python
df_fix['home'] = df_fix['home'].str.strip()
df_fix['away'] = df_fix['away'].str.strip()
```

```python
 df_hist['score'].str.split('–', expand = True)
```

```python
df_hist[['HomeGoals','AwayGoals']]=  df_hist['score'].str.split('–', expand = True)
```

```python
df_hist
```

```python
df_hist.drop('score', axis=1, inplace=True)
```

```python
df_hist.dtypes
```

```python
df_hist.rename(columns={'home': 'HomeTeam', 'away': 'AwayTeam', 'year': 'Year'}, inplace= True)
```

```python
df_hist[df_hist['AwayGoals'].str.contains('[^\d–]')]
```

```python
df_hist['AwayGoals']= df_hist['AwayGoals'].str.replace('[^\d–]', '', regex= True)
```

```python
df_hist=df_hist.astype({'HomeGoals': int, 'AwayGoals':int, 'Year': int})
```

```python
df_hist
```

```python
df_hist['Totalgoal'] = df_hist['HomeGoals']+ df_hist['AwayGoals']
```

```python
df_hist
```

```python
df_hist.to_csv('clean_fifa_worldcup_matches.csv', index = False)
df_fix.to_csv('clean_fifa_worldcup_fixture.csv', index = False)
```
